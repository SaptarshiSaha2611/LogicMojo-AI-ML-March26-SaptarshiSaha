Context: Before training any ML model, we analyze feature distributions.

Mean
Median
Standard deviation
Variance (for each feature)
3. Identify:

```python
import numpy as np
from sklearn.datasets import load_iris

# Load Iris 
iris = load_iris()
X = iris.data  

# 1. Stats calculation (column-wise for the 4 features)
means = np.mean(X, axis=0)
medians = np.median(X, axis=0)
stds = np.std(X, axis=0)
vars = np.var(X, axis=0)

print(f"Features: {iris.feature_names}")
print(f"Mean: {means}\nMedian: {medians}\nStd Dev: {stds}\nVariance: {vars}")

# 2. Identify highest variability
high_var_idx = np.argmax(vars)
print(f"\nHighest variability feature: {iris.feature_names[high_var_idx]} ({vars[high_var_idx]:.2f})")

# 3. Reshape petal length (index 2) to (n, 1)
petal_len_2d = X[:, 2].reshape(-1, 1)
print(f"Reshaped petal length from {X[:, 2].shape} to {petal_len_2d.shape}")

"""
Quick Notes:
- Variability: Petal length usually has the highest variance. In ML, high-variance features 
  often carry more 'signal' for classification, but can also dominate models (like KNN) 
  if not scaled properly.
- Shape (n, 1): Most Scikit-Learn estimators expect a 2D array (samples, features). 
  Even if you only have one feature, the model needs that second dimension to 
  understand that it's looking at a single 'column' of data.
"""
```

Q2. Data Selection for Model Input

Context: Models are trained on selective features, not full raw data.

Dataset: Iris

Tasks:

1. Extract:
First 100 samples
Only last 2 features
2. Use boolean masking:
Select samples where petal length is greater than dataset mean
3. Count selected samples
4. Explain how this relates to feature-based filtering in ML

```python
import numpy as np 
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
X = iris.data

# 1. Slice: First 100 rows and only the last 2 columns (Petal Length/Width)
subset = X[:100, -2:]

# 2. Boolean Masking: Get rows where petal length > mean
# Since we sliced the last 2 columns, petal length is now index 0
petal_len = subset[:, 0]
mean_val = np.mean(petal_len)

# Apply mask and filter
mask = petal_len > mean_val
filtered_data = subset[mask]

# 3. Count results
count = len(filtered_data)

print(f"Threshold (Mean): {mean_val:.2f}")
print(f"Samples exceeding mean: {count}")

"""
ML Context: Feature-based Filtering
- Data Quality: Filtering helps remove outliers or irrelevant noise that can confuse a model.
- Class Separation: In the Iris case, petal dimensions are highly discriminative. By 
  subsetting based on these values, we are essentially identifying the 'thresholds' 
  that a decision tree or linear model would use to separate species.
- Efficiency: Dropping features/rows that don't meet specific criteria reduces 
  computational load and prevents 'overfitting' to insignificant data points.
"""
```

```python
import numpy as np 
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
X = iris.data

# 1. Standardization 

mu = X.mean(axis=0)
sigma = X.std(axis=0)

X_scaled = (X - mu) / sigma

# 2. Verify results
scaled_mu = X_scaled.mean(axis=0)
scaled_std = X_scaled.std(axis=0)

print("--- Scaling Verification ---")
print(f"Centering (Mean ≈ 0): {np.round(scaled_mu, 2)}")
print(f"Scaling   (Std  ≈ 1): {np.round(scaled_std, 2)}")
```

Q6. Handling Missing Data (Critical ML Step)

Context: Models cannot handle null values.

Dataset: Titanic

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column
4. Explain:
Why different strategies are used for different columns

```python
import seaborn as sns
import pandas as pd

# Load data
df = sns.load_dataset('titanic')

# 1. Imputation: Age
df['age'] = df['age'].fillna(df['age'].median())

# 2. Imputation: Embarked
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

# 3. Dimensionality Reduction: Drop Deck
df.drop(columns=['deck'], inplace=True)

# Final check
print("Remaining Nulls:\n", df.isnull().sum())
print(f"\nFinal dataset shape: {df.shape}")

"""
Strategic Rationale for Missing Data:
- Numeric vs. Categorical: We use the median for 'Age' because it's a continuous variable 
  where extreme values (outliers) could skew a simple average. For 'Embarked', since 
  it's categorical, 'mean' doesn't exist, so we use the most common port (mode).
- Information Loss vs. Noise: We fill 'Age' and 'Embarked' because they are critical 
  features. However, for 'Deck', the missingness is so high that any guess (imputation) 
  would introduce more noise than actual signal, so dropping it is the safer choice.
- Model Requirement: Most scikit-learn models will crash if they encounter a 'NaN'. 
  Handling these now ensures the pipeline is robust before we start training.
"""
```

Q7. Feature Engineering for Model Improvement

Context: Raw data is rarely useful—features must be created.

Dataset: Titanic

Tasks:

1. Create:
family_size = sibsp + parch + 1
2. Create:
is_alone (binary feature)
3. Compute survival rate by is_alone
4. Explain:
Why engineered features improve ML models

```python
import seaborn as sns
import pandas as pd

# Load and prep data
df = sns.load_dataset('titanic')

# Feature Engineering: 
# Combining family counts and flagging solo travelers
df['family_size'] = df['sibsp'] + df['parch'] + 1
df['is_alone'] = (df['family_size'] == 1).astype(int)

# Check survival rates to validate the new features
survival_by_alone = df.groupby('is_alone')['survived'].mean().to_frame('Survival Rate')

print("Analysis of 'is_alone' impact on survival:")
print(survival_by_alone)

"""
Quick Breakdown: Why this helps the model
- Consolidates Features: 'family_size' merges two related columns (sibsp/parch) into one clearer signal.
- Flags Specific States: The jump from 1 family member to 0 is often a 'tipping point' for survival that models might miss if left as raw counts.
- Better Signal: It filters out noise; the binary 'is_alone' state is often more predictive than the exact number of relatives.
- Context: It helps the model pick up on historical protocols like 'women and children first' which often favored families over solo men.
"""
```

Q8. Group-Based Insights (Customer Segmentation Logic)

Context: ML models often depend on aggregated patterns.

Dataset: Tips

Tasks:

1. Compute:
Average tip by day
Total bill by gender
2. Create new feature:
tip_percentage
3. Identify:
Which day has highest tipping behavior
4. Explain:
How this helps in segmentation models

```python
import seaborn as sns
import pandas as pd

# Load dataset
tips = sns.load_dataset('tips')

# 1. Feature Engineering: tip_percentage
# 
tips['tip_pct'] = (tips['tip'] / tips['total_bill']) * 100

# 2. 
# Average tip by day
avg_tip_day = tips.groupby('day')['tip'].mean().to_frame('Avg Tip')

# Total bill by gender
total_bill_gender = tips.groupby('sex')['total_bill'].sum().to_frame('Total Revenue')

# 3. Identify highest tipping day 

best_day = tips.groupby('day')['tip'].sum().idxmax()
max_bill = tips.groupby('day')['tip'].sum().max()

print("--- Group Statistics ---")
print(avg_tip_day)
print(f"\n{total_bill_gender}")
print(f"\nBest tipping day: {best_day} {max_bill:.2f}")

"""
How this helps in Segmentation Models:
- Behavioral Profiling: Features like 'tip_pct' allow the model to group customers 
  by 'generosity' rather than just spending power. A person with a small bill who 
  tips 30% is a different segment than a big spender who tips 10%.
- Temporal Trends: Identifying the 'highest tipping day' helps a model account for 
  time-based bias. If the model knows it's Sunday, it can adjust survival or 
  spending expectations based on historical group averages.
- Feature Scaling: Aggregates help in 'Target Encoding'. Replacing a category 
  (like 'Day') with its average tip value helps mathematical models better 
  understand the influence of time on the target variable.
"""
```

Q9. Advanced Filtering + Subsetting

Context: Training data is often filtered based on business rules.

Dataset: Tips

Tasks:

1. Filter:
total_bill > 20 AND tip < 3
2. Select only relevant columns
3. Analyze:
Is this segment under-tipping?

```python
import seaborn as sns
import pandas as pd

# Load dataset
tips = sns.load_dataset('tips')

# 1. Advanced Filtering: High bills (> 20) but low tips (< 3)
under_tippers = tips.loc[
    (tips['total_bill'] > 20) & (tips['tip'] < 3), 
    ['total_bill', 'tip', 'sex', 'day', 'size']
]

# 2. Analysis: Comparison logic

avg_pct = (tips['tip'] / tips['total_bill']).mean()
seg_pct = (under_tippers['tip'] / under_tippers['total_bill']).mean()

print(f"Global Average Tip: {avg_pct:.2%}")
print(f"Segment Average Tip: {seg_pct:.2%}")
print(f"Analysis: The segment is tipping {avg_pct - seg_pct:.2%} less than average.")

"""
How this helps Anomaly Detection:
- Isolating Noise: By filtering for high bills with low tips, we isolate 'anomalous' 
  records that break the typical linear relationship between bill size and tip.
- Data Quality: In a real ML pipeline, these might represent data entry errors or 
  extreme outliers that could skew a regression model's coefficients.
- Fraud/Pattern Detection: In business logic (like banking or retail), this type 
  of subsetting is used to flag transactions that fall outside of 'Expected 
  Behavioral Norms,' which is the core of supervised anomaly detection.
"""
```

Q10. Data Transformation for Model Input (NumPy)

Context: ML models require structured numeric arrays.

Tasks:

1. Create two arrays of shape (3,2)
2. Perform:
Vertical stacking
Horizontal stacking
3. Reshape into (2,6)
4. Explain:
Why reshaping is required in ML pipelines

```python
import numpy as np

# 1. Create two arrays of shape (3, 2)
arr1 = np.array([[1, 2], [3, 4], [5, 6]])
arr2 = np.array([[7, 8], [9, 10], [11, 12]])

# 2. Stacking operations
v_stack = np.vstack((arr1, arr2))

# Horizontal stacking
h_stack = np.hstack((arr1, arr2))

# 3. Reshape into (2, 6)
reshaped_arr = v_stack.reshape(2, 6)

print("Vertical Stack Shape:", v_stack.shape)
print("Horizontal Stack Shape:", h_stack.shape)
print("Final Reshaped Array (2, 6):")
print(reshaped_arr)

"""
Why Reshaping is required in ML Pipelines:
- Input Alignment: Most ML models (like Scikit-Learn's .fit()) expect a 2D array 
  of (n_samples, n_features). If your data is 1D or 3D, you must reshape it to 
  match the model's expected 'input_shape'.
- Neural Network Layers: In Deep Learning (CNNs), we often 'flatten' multi-dimensional 
  images into a 1D vector before passing them into a fully connected layer.
- Compatibility: When merging datasets or performing matrix multiplication (dot 
  products), the inner dimensions must align. Reshaping ensures the math is valid 
  without changing the underlying data.
"""
```

Q11. End-to-End ML Preprocessing Pipeline (Titanic)

Context: Prepare dataset for ML model training

Tasks:

1. Load dataset
2. Handle missing values
3. Encode categorical variables (sex, embarked)
4. Normalize numerical features
5. Convert to NumPy array

Output:

Final feature matrix ready for model training

```python
import seaborn as sns
import pandas as pd
import numpy as np

# 1. Load Data
df = sns.load_dataset('titanic')

# 2. Handle Missing Values
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

# Drop high-null and redundant columns
df.drop(columns=['deck', 'embark_town', 'class', 'adult_male'], inplace=True)

# 3. Encode Categorical Variables
df = pd.get_dummies(df, columns=['sex', 'embarked'], drop_first=True)

# 4. Normalize Numerical Features (Min-Max Scaling)

cols_to_norm = ['age', 'fare']
df[cols_to_norm] = (df[cols_to_norm] - df[cols_to_norm].min()) / (df[cols_to_norm].max() - df[cols_to_norm].min())

# 5. Convert to NumPy Array
# X = Features, y = Target
X = df.drop(columns=['survived']).values
y = df['survived'].values

# Verification
print(f"Feature Matrix (X) Shape: {X.shape}")
print(f"Target Array (y) Shape: {y.shape}")
print("\nFirst row of processed data:")
print(X[0])
```

Q12. Feature Selection using Statistical Understanding

Context: Not all features improve model performance.

Dataset: Iris

Tasks:

1. Compute correlation matrix
2. Identify highly correlated features
3. Drop redundant features
4. Explain:
Impact on overfitting and model performance

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('iris')
print("Step 1: Raw Data Sample")
print(df.head(3))

# 2. Compute Correlation Matrix
corr_matrix = df.select_dtypes(include=['number']).corr()

print("\nStep 2: Correlation Matrix")
print(corr_matrix.round(2))

# Unstacking turns the grid into a single column of pairs
high_corr = corr_matrix.unstack().sort_values(ascending=False)

# Filter out the 1.0 values 
high_corr = high_corr[high_corr < 1]

print("\nStep 3: Top Correlated Pairs")
print(high_corr[high_corr > 0.85])

# Dropping the feature that provides the same info as another
df_reduced = df.drop(columns=['petal_width'])

print("\nStep 4: New Column List")
print(df_reduced.columns.tolist())

print(f"\nOriginal features: {df.columns.tolist()}")
print(f"Reduced features:  {df_reduced.columns.tolist()}")

"""
Impact on Model Performance & Overfitting:
- Reducing Overfitting: Highly correlated features (Multicollinearity) can cause 
  the model to become over-sensitive to small fluctuations in the training data. 
  By dropping the redundant 'petal_width', we force the model to generalize better.
- Occam's Razor: Simpler models are usually more robust. Removing redundancy 
  reduces 'Degrees of Freedom', making the model less likely to 'memorize' noise.
- Computational Efficiency: Fewer features mean faster training and lower 
  memory usage, which is critical when scaling to millions of rows.
- Interpretability: It’s easier to explain a model to stakeholders when every 
  feature provides a unique, distinct piece of information.
"""
```

Child / Adult / Senior

4. Interpret:

```python
import seaborn as sns 
import pandas as pd 

df = sns.load_dataset('titanic')
df.head()
```

```python
import seaborn as sns
import pandas as pd

# 1. Load dataset
df = sns.load_dataset('titanic')

# 2. Define Custom Logic
def categorize_age(age):
    if age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

# 3. Use .apply() to create the new feature
df['age'] = df['age'].fillna(df['age'].median())
df['age_group'] = df['age'].apply(categorize_age)

# 4. Compute Survival Rate per group
survived_rate = df.groupby('age_group')['survived'].mean().sort_values(ascending=False)

print("Survival Rates by Age Group:")
print(survived_rate)


```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

# 1. Generate 1,000 random error values (Normal Distribution)
# We simulate a model that is mostly accurate but has some noise
np.random.seed(42) # For reproducibility
errors = np.random.normal(loc=0, scale=5, size=1000)

# 2. Compute Mean and Standard Deviation
mean_error = np.mean(errors)
std_error = np.std(errors)

# 3. Identify Outliers: Mean ± 2 * Std
# This range covers roughly 95% of data in a normal distribution
lower_bound = mean_error - (2 * std_error)
upper_bound = mean_error + (2 * std_error)

outliers = errors[(errors < lower_bound) | (errors > upper_bound)]

print(f"Mean Error: {mean_error:.4f}")
print(f"Standard Deviation: {std_error:.4f}")
print(f"Number of Outliers: {len(outliers)}")
print(f"Outlier Thresholds: {lower_bound:.2f} to {upper_bound:.2f}")

"""
What High Variance means in Model Performance:
- Inconsistent Predictions: High variance (High Std Dev) means the model's 
  errors are widely spread. One prediction might be off by 1, and the next 
  by 50. The model is 'unreliable.'
- Overfitting: Often, high variance is a sign that the model has 'memorized' 
  the training noise. It works perfectly on known data but fails wildly on 
  new, unseen data.
- Reliability: In a business context (like predicting house prices), a model 
  with high variance is dangerous because you cannot trust its confidence 
  intervals. You would prefer a slightly less accurate model that is 
  consistently 'close' over one that is sometimes perfect and sometimes 
  terribly wrong.
"""
```
