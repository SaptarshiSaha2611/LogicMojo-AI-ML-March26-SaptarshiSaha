Q1. Division with Exception Handling

Write a program that takes two numbers and safely performs division.
Handle division by zero using try-except.

````python
def safe_division():
    try:
       
        numerator = float(input("Enter the numerator: "))
        denominator = float(input("Enter the denominator: "))
        
     
        result = numerator / denominator
        print(f"Result: {result}")
        
    except ZeroDivisionError:
        print("Error: You cannot divide by zero")
    except ValueError:
        print("Error: Please enter valid numerical values.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


safe_division()
````

Skip invalid values using exception handling.

Q16. Extract Dates

Extract date from:

text = "Today is 2026-03-28"

Return:

Year
Month
Day

````python
import re

text = "Today is 2026-03-28"

pattern = r"(\d{4})-(\d{2})-(\d{2})"

match = re.search(pattern, text)

if match:
    year = match.group(1)
    month = match.group(2)
    day = match.group(3)
    
    print(f"Year: {year}")
    print(f"Month: {month}")
    print(f"Day: {day}")
````

## 1. Customer Segmentation Engine

You are given a list of customer dictionaries:

```python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
```

Task:

1. Write a function segment_customer(customer):
2. If spend divided by income is greater than 0.6 return "High Value"
3. If spend divided by income is between 0.3 and 0.6 return "Medium Value"
4. Otherwise return "Low Value"
5. Create a final dictionary in the format {customer_id: segment}

````python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]

def segment_customer(customer):
    
    ratio = customer["spend"] / customer["income"]
    
    if ratio > 0.6:
        return "High Value"
    elif 0.3 <= ratio <= 0.6:
        return "Medium Value"
    else:
        return "Low Value"


segmentation_results = {c["id"]: segment_customer(c) for c in customers}

print(f"Customer Segments: {segmentation_results}")
````

## 2. Text Cleaning Pipeline

You are given:

```python
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
```

Task:

1. Write a function clean_text(text):
2. Remove special characters
3. Remove extra spaces
4. Convert text to lowercase
5. Apply the function to all elements using loop or map
6. Return cleaned list

---

````python
import re

texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]

def clean_text(text):
    # 1. Remove special characters 

    text = re.sub(r'[^a-zA-Z0-9 ]', ' ', text)
    
    # 2. Convert to lowercase
    text = text.lower()
    
    # 3. Remove extra spaces 
  
    text = " ".join(text.split())
    
    return text


cleaned_list = [clean_text(t) for t in texts]


print(f"Cleaned:  {cleaned_list}")
````

## 4. Feature Engineering

```python
data = [

```

1. Write a function feature_engineer(record)
2. Add a new key salary_per_exp
3. salary_per_exp is salary divided by experience
4. Handle division by zero using conditional logic
5. Return updated dataset

---

````python
data = [
    {"salary": 50000, "experience": 2},
    {"salary": 100000, "experience": 5},
]

def feature_engineer(record):

    if record["experience"] > 0:
  
        record["salary_per_exp"] = record["salary"] / record["experience"]
    else:

        record["salary_per_exp"] = record["salary"]
    
    return record


updated_data = [feature_engineer(item) for item in data]


for entry in updated_data:
    print(entry)
````

## 5. Custom Reduce Function

You are given:

```python
accuracies = [0.8, 0.75, 0.9, 0.85]
```

Task:

1. Use reduce function to calculate average accuracy
2. Do not use sum function
3. Show intermediate logic using a custom function


````python
from functools import reduce

accuracies = [0.8, 0.75, 0.9, 0.85]

def accumulate_total(current_total, next_value):
    
    new_total = current_total + next_value
    print(f"Adding {next_value:^6} | New Running Total: {new_total:.2f}")
    return new_total


total_sum = reduce(accumulate_total, accuracies)


average_accuracy = total_sum / len(accuracies)


print(f"Final Average Accuracy: {average_accuracy:.2f}")
````

## 6. Sequence Pattern Detection

You are given:

```python
data = [10, 20, 30, 25, 40, 50]
```

Task:

1. Write a function detect_increasing_sequence(data)
2. Find the longest continuous increasing sequence
3. Return that sequence as a list

Example output:

```python
[10, 20, 30]
```


````python
def detect_increasing_sequence(data):
    if not data:
        return []

    longest_seq = []
    current_seq = [data[0]]

    # Start from the second element
    for i in range(1, len(data)):
        if data[i] > data[i-1]:
            current_seq.append(data[i])
        else:
          
            if len(current_seq) > len(longest_seq):
                longest_seq = current_seq
       
            current_seq = [data[i]]


    if len(current_seq) > len(longest_seq):
        longest_seq = current_seq

    return longest_seq

data = [10, 20, 30, 25, 40, 50]
result = detect_increasing_sequence(data)


print(f"Longest Increasing Sequence: {result}")
````

## 7. Dictionary-Based Recommendation System

You are given:

```python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
```

Task:

1. Write a function recommend(user)
2. Recommend items based on common items between users
3. Do not recommend items already purchased by the user
4. Use sets and dictionary iteration

---

````python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}

def recommend(target_user):
    # 1
    target_items = set(user_purchases[target_user])
    recommendations = set()

   
    for other_user, other_items in user_purchases.items():
        if other_user == target_user:
            continue  
        
        other_items_set = set(other_items)
        

        common_items = target_items.intersection(other_items_set)
        
        if common_items:

            new_suggestions = other_items_set.difference(target_items)
            recommendations.update(new_suggestions)

    return list(recommendations)


print(f"Recommendations for user1: {recommend('user1')}")
print(f"Recommendations for user2: {recommend('user2')}")
````

## 8. Recursive Decision Tree Simulation

Task:

1. Write a recursive function decision_tree(score)
2. If score is greater than 80 return "Approved"
3. If score is between 50 and 80 call the function again with score plus 10
4. If score is less than 50 return "Rejected"

---


````python
def decision_tree(score):
    print(f"Processing score: {score}")  
    

    if score > 80:
        return "Approved"
    
 
    elif 50 <= score <= 80:
        print("Result: Borderline. Re-evaluating with +10 points.")
        return decision_tree(score + 10)
    

    else:
        return "Rejected"


print(f"Final Decision: {decision_tree(65)}")
print("-" * 20)
print(f"Final Decision: {decision_tree(45)}")
````
